```python
x= data.drop(['Purchased'],axis=1)
```

```python
x
```

```python
y= data['Purchased']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.25)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(n_estimators=200)
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_test,y_test)
```

```python
y_pred = clf.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
new_data = [input("Enter your Age : "), input("\nEnter Your estimated salary : ")]
result = clf.predict([new_data])

print("Predicted Value is ", result[0])

if result[0]==1:
  print("The individual will purchase ")
else:
  print("The individual will not purchase ")
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(15, 10))
plot_tree(clf.estimators_[0], filled=True, rounded=True, class_names=['Not Purchased', 'Purchased'])
plt.show()
```
